```python
def dimensoesObjeto(): # Função que calcula as dimensões e retorna o valor do volume
  while True: # Laço de repetição 
    try: # Tenta validar todos os dados
      #Variáveis
      altura = float(input('Digite a altura do objeto em cm: ')) # Variável local de altura
      comprimento = float(input('Digite o comprimento do objeto em cm: ')) # Variável local de comprimento
      largura = float(input('Digite a largura do objeto em cm: ')) # Variável local de largura
      volume = altura * comprimento * largura # Variável local de volume multiplicando altura x comprimento x largura
      global valor # Variável global do valor que será usado fora da função
      
      if  volume < 1000:
        valor = 10 # Se o volume for menor que 1000, o valor será 10
      elif 1000 <= volume < 10000:
        valor = 20 # Se o volume for menor ou igual a 1000 e menor que 10000, o valor será 20
      elif 10000 <= volume < 30000:
        valor = 30 # Se o volume for menor ou igual a 10000 e menor que 30000, o valor será 30
      elif 30000 <= volume < 100000:
        valor = 50 # Se o volume for menor ou igual a 30000 e menor que 100000, o valor será 50
      elif volume >= 100000:
        print(f'O volume de {volume} não é aceito. Digite as dimensões novamente.') # Se o volume for maior que 100000 não será aceito
        continue # Volta ao começo se for o volume for maior que 100000
    
    except: # Capta o erro se o valor não for numérico
      print('Este valor não é numérico. Tente Novamente.') # Print a mensagem se o valor não for númerico
      continue # Volta ao começo se for o volume não for númerico
    
    print(f'O volume é de {volume}cm³ (altura {altura}cm, comprimento {comprimento}cm, largura {largura})cm.\n') # Print das dimensões
    
    return valor # Valor a ser retornado

def pesoObjeto(): # Função que retorna o peso
  while True: # Laço de repetição
    try: # Tenta validar todos os dados
      # Variáveis
      peso = float(input('Digite o peso do objeto em kg: ')) # Variável local do peso
      global multiplicadorPeso # Variável global do multiplicador do peso que será usado fora da função
      
      if peso <= 0.1:
        multiplicadorPeso = 1  # Se o peso for menor ou igual a 0.1, o multiplicador será 1
      elif 0.1 < peso <= 1:
        multiplicadorPeso = 1.5 # Se o peso for menor que 0.1 e menor ou igual a 1, o multiplicador será 1.5
      elif 1 < peso <= 10:
        multiplicadorPeso = 2 # Se o peso for menor que 1 e menor ou igual a 10, o multiplicador será 2
      elif 10 < peso <= 30:
        multiplicadorPeso = 3 # Se o peso for menor que 10 e menor ou igual a 30, o multiplicador será 3
      elif peso > 30:
        print(f'{peso}kg não é aceito. Digite o outro peso.') # Se o peso for maior que 30 não será aceito
        continue # Volta ao começo se for o peso for maior que 30
    
    except: # Capta o erro se o valor não for numérico
      print('Este valor não é numérico. Tente novamente.') # Print a mensagem se o valor não for númerico
      continue # Volta ao começo se o peso não for númerico
    
    print(f'O peso é de {peso}kg.\n') # Print do peso digitado
    
    return multiplicadorPeso # Retorna o multiplicador do peso

def rotaObjeto():
  while True: # Laço de repetição
    # Print das opcões de rota disponiveis 
    print('RS - De Rio de Janeiro até São Paulo')
    print('SR - De São Paulo até Rio de Janeiro')
    print('BS - De Brasília até São Paulo')
    print('SB - De São Paulo até Brasília')
    print('BR - De Brasília até Rio de Janeiro')
    print('RB - De Rio de Janeiro até Brasília')
    # Variáveis
    rota = input('Digite a rota desejada: ') # Variável local da rota
    global multiplicadorRota # Variável global do multiplicador da rota que será usado fora da função
      
    if rota == 'RS':
      print('A rota escolhida foi RS - De Rio de Janeiro até São Paulo.\n') # Print da rota escolhida
      multiplicadorRota = 1 # Se a rota for RS, o multiplicador será 1
      break # Termina o laço quando a rota for escolhida
    elif rota == 'SR':
      print('A rota escolhida foi SR - De São Paulo até Rio de Janeiro.\n') # Print da rota escolhida
      multiplicadorRota = 1 # Se a rota for SR, o multiplicador será 1
      break # Termina o laço quando a rota for escolhida
    elif rota == 'BS':
      print('A rota escolhida foi BS - De Brasília até São Paulo.\n') # Print da rota escolhida
      multiplicadorRota = 1.2 # Se a rota for BS, o multiplicador será 1.2
      break # Termina o laço quando a rota for escolhida
    elif rota == 'SB':
      print('A rota escolhida foi SB - De São Paulo até Brasília.\n') # Print da rota escolhida
      multiplicadorRota = 1.2 # Se a rota for SB, o multiplicador será 1.2
      break # Termina o laço quando a rota for escolhida
    elif rota == 'BR':
      print('A rota escolhida foi BR - De Brasília até Rio de Janeiro.\n') # Print da rota escolhida
      multiplicadorRota = 1.5 # Se a rota for BR, o multiplicador será 1.5
      break # Termina o laço quando a rota for escolhida
    elif rota == 'RB':
      print('A rota escolhida foi RB - De Rio de Janeiro até Brasília.\n') # Print da rota escolhida
      multiplicadorRota = 1.5 # Se a rota for RB, o multiplicador será 1.5
      break # Termina o laço quando a rota for escolhida
    else:
      print('Essa Rota não existe. Digite uma rota existente.\n') # Print se a rota não existir
      continue # Volta ao começo se a rota não existir
    
  return multiplicadorRota # Retorna o multiplicador da rota

print('Olá, seja bem-vindo ao programa de logística do Diarley Silva Santos (RU 4095617).\n')

# Chamando as funcões de dimensão, peso e rota
dimensoesObjeto()
pesoObjeto()
rotaObjeto()

total = valor * multiplicadorPeso * multiplicadorRota # Total a ser pago multiplicando o valor pelo multiplicador do peso e rota
print(f'O valor é de R${valor} * Peso {multiplicadorPeso} * Rota {multiplicadorRota}.') # Valores separados
print(f'Total a ser pago é: R${total}') # Total a ser pago
```
